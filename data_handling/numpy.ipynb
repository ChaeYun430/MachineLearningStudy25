```python
import numpy as np

# NumPy
# array(), astype(), arange(), zeros(), ones(), reshape()

# N-dimensional array
# shape, ndim, dtype
# ndarray 내의 데이터 타입은 연산의 특성상 같은 데이터 타입만 가능
# 만약 리스트를 ndarray으로 변경시 데이터 크기가 더 큰 데이터 타입으로 형 변환을 일괄 적용한다.

array1 = np.array([1, 2, 3])
print(type(array1))
print(array1.shape)
array2 = np.array((['A', 2, 3], [1, 2, 3], [1, 2, 3], [1, 2, 3]))
print(type(array2))
print(array2.shape)
print('{0}차원, {1}차원'.format(array1.ndim, array2.ndim))
print('{0}타입, {1}타입'.format(array1.dtype, array2.dtype))
print('{}타입'.format(array1.astype('int8').dtype)) # 인자로 원하는 타입을 문자열로 지정

sequence_array = np.arange(10)
print(sequence_array)
print(sequence_array.dtype, sequence_array.shape)
# 인자로 튜플 형태의 shape를 입력하면 모든 값을 0/1로 채운 객체 반환
zero_array = np.zeros((2, 5), dtype = 'int8')
print(zero_array)
print(zero_array.dtype, zero_array.shape)
one_array = np.ones((5, 2))
print(one_array)
print(one_array.dtype, one_array.shape)

# 특정 차원 및 크기로 변환
# 지정된 사이즈로 변경 불가시 오류 발생
# -1을 인자로 사용하면 원래 ndarray와 호환되는 새로운 shape로 변환
# 열(행)에 맞는 행(열)을 자동으로 새롭게 생성해 변환하라
arrayA = np.arange(10)
print(arrayA, arrayA.shape)
arrayB = arrayA.reshape(2, 5)
print(arrayB, arrayB.shape)
arrayC = arrayB.reshape(5, -1)
print(arrayC, arrayC.shape)
arrayD = np.arange(8).reshape(2, 2, 2)
print(arrayD.tolist(), arrayD.shape)
arrayE = arrayD.reshape(-1, 1)
print(arrayE.tolist(), arrayE.shape)
```

```python
# ===== Indexing =====
import numpy as np
array = np.arange(1, 10)
array1 = array.reshape(3, 3)

# ----- 특정 데이터 추출 ------
print(array1)
print('axis 0 : [1], axis 1 : [0], {} \n'.format(array1[1, 0]))

# ----- Slicing -----
print(array1[1:3, 0:3])
print(array1[1:3, :3])
print(array1[1:3, :])
print(array1[1])

# ----- Fancy Indexing -----
# list나 ndarray로 인덱스 집합을 지정하면 해당 위치의 인덱스에 해당하는 ndarray 반환
# 행렬 >> 인덱스  / 리스트와 슬라이싱 포함관계 주의하기
print(array1[[0,2], 2])
print(array1[[0,2], 0:2])
print(array1[[0,2]])

# ----- Boolean Indexing -----
print(array1 > 5)
# 조건 입력시 반환된 ndarray 객체를 []내에 입력하면 True 값이 있는 위치 인덱스 값으로 자동 변환해 해당하는 인덱스 위치의 데이터가 반환된다.
print(array1[array1 > 5])
```

```python
# ===== 행렬의 정렬 =====
import numpy as np

# ----- 1차원 ------
org_array = np.array([3, 1, 9, 5])

# 비파괴적 처리
print(np.sort(org_array))
print(org_array)

# 파괴적 처리
print(org_array.sort())
print(org_array)

# 내림차순
print(np.sort(org_array)[::-1])

# ----- 2차원 -----
org_array_2d = np.array([[8, 12], [7, 1]])
print(org_array_2d)
print(np.sort(org_array_2d, axis = 0))
print(np.sort(org_array_2d, axis = 1))

# ----- 정렬된 행렬의 원본 행렬 인덱스 반환 -----
print(np.argsort(np.array([3, 1, 9, 5])))
print(np.argsort(np.array([3, 1, 9, 5]))[::-1])
# 반환값을 fancy index에 적용해 메타데이터와 같이 활용 가능
```

```python
# ===== 선형대수 연산 =====

import numpy as np
# dot(), transpose()

array1 = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
array2 = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
print(array1)
print(array2)

# ----- 행렬 내적 -----
# 왼쪽 행렬의 열 개수와 오른쪽 행렬의 행의 개수가 동일해야 내적 연산이 가능함.
print(np.dot(array1, array2))

# ----- 전치 행렬 -----
# 기존 행렬 시계방향 90도 회전, 좌우반전
print(np.transpose(array1))
print(np.transpose(array2))
```
