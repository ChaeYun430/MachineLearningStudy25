```python
import matplotlib.pyplot as plt
# plot(), title(), show(), xticks()
import numpy as np


plt.plot(np.arange(-1, 100), 2*np.arange(-1, 100))
plt.title('Hello pyplot')
plt.xticks(ticks=np.arange(0,100,5), rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
# figure(), axes()

# <class 'matplotlib.axes._axes.Axes'>
# plot(), set_title(), set_xlabel(), set_facecolor()

fig, ax = plt.subplots(figsize=(5, 4), facecolor='skyblue')
ax.set_title("Hello pyplot(figure, axes)")
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_facecolor('white')

ax.plot([1, 2, 3], [3, 6, 9], color='green',  marker='*', linestyle='dashed', linewidth=2, markersize=10) # X, Y좌표 원소의 개수가 같아야 한다.
print(type(fig), type(ax))
```

```python
# ===== 동일 축 그래프 (multiple plots on same axes) & 범례 =====
import numpy as np
import matplotlib.pyplot as plt

x_value_01 = np.arange(0, 100)
y_value_01 = 2 * x_value_01
y_value_02 = 4 * x_value_01

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Hello plot')

plt.plot(x_value_01, y_value_01, color = 'red', label = 'red label')
plt.plot(x_value_01, y_value_02, color = 'green', label = 'green label')
plt.legend()
```

```python
# ===== line & bar chart =====
import matplotlib.pyplot as plt
import numpy as np

x_value_01 = np.arange(0, 10)
y_value_01 = 2 * x_value_01

fig, ax = plt.subplots(figsize=(5, 4), facecolor='skyblue')
ax.set_title("Hello pyplot(figure, axes)")
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_facecolor('white')

ax.plot(x_value_01, y_value_01, color='red',  marker='o', linestyle='dashed', linewidth=2, markersize=5, label = 'line plot')
ax.bar(x_value_01, y_value_01, color = 'green', label = 'bar plot')
plt.legend()
plt.show()
```

```python
# ===== subplot별로 시각화 =====
import matplotlib.pyplot as plt
import numpy as np

# fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (6, 3))
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (8, 6), facecolor = 'skyblue')
print(type(ax), type(ax[0]))

x_value_01 = np.arange(0, 10)
y_value_01 = 2 * x_value_01
y_value_02 = 4 * x_value_01

ax[0][0].plot(x_value_01, y_value_01, color='red',  marker='o', linestyle='dashed', linewidth=2, markersize=5, label = 'line plot')
ax[0][0].bar(x_value_01, y_value_01, color = 'green', label = 'bar plot')
ax[0][1].plot(x_value_01, y_value_01, color = 'red', label = 'red label')
ax[0][1].plot(x_value_01, y_value_02, color = 'green', label = 'green label')

ax[0][0].set_title('[0][0]')
ax[0][0].legend()
ax[0][1].set_title('[0][1]')
ax[0][1].legend()
```
