```python
# ===== Limit of Accuracy =====
from sklearn.datasets import load_digits
# MINIST  데이터 세트는 0부터 9까지의 숫자 이미지의 픽셀 정보를 가지고 있으며,
#  이를 기반으로 숫자 Digit을 예측하는데 사용
from  sklearn.model_selection import train_test_split
from sklearn.base import BaseEstimator
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd

class CustomClassifier(BaseEstimator):
  def fit(self,  X, y):
    pass

  def predict(self, X): # 입력값에 상관없이 0으로 예측
    return np.zeros((len(X), 1), dtype = bool)

digits = load_digits()
y = (digits.target == 7).astype(int) # 데이터셋의 레이블을 이진 분류용으로 변환
X_train, X_test, y_train, y_test = train_test_split(digits.data, y)

custom_clf = CustomClassifier()
custom_clf.fit(X_train, y_train)
pred = custom_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(accuracy)
```

```python
# ===== Confusion matrix =====
# 이진 분류의 예측  오류가 얼마인지와 더불어 어떠한 유형의 예측 오류가 발생하고 있는지를 함께 나타내는 지표
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred)
# 중점적으로 찾아야 하는 매우 적은 수의 결괏값에 Positive를 설정한다.
# 정확도 지표는 비대칭한 데이터 세트에서 Positive에 대한 예측 정확도를 판단하지 못한 채
# Negative에 대한 예측 정확도만으로도 분류의 정확도가 매우 높게 나타나는 수치적인 판단 오류를 일으키게 된다.
```
